```python
import numpy as np
import pandas as pd

path = 'HelmLiteHeatmap'
#alternatives:
    #'HelmLiteHeatmap_0.9_0.1'
    #'HelmLiteHeatmap_0.1_0.4'

df = pd.read_pickle(f'../results/{path}/heatmap_data_compressed.pkl')

M = df.pivot(index='source', columns='target', values='auroc')
M = M.fillna(M.median(axis=1), axis=0)
models = M.index

mean_as_source = M.mean(axis=1)
mean_as_target = M.mean(axis=0)
in_domain = pd.Series(np.diag(M.values), index=models)
best_target_name = M.idxmax(axis=1)
best_target_value = M.max(axis=1)
best_source_name = M.idxmax(axis=0)  
best_source_value = M.max(axis=0)

summary = pd.DataFrame({
    "name": models,
    "mean_as_source": mean_as_source.values,
    "mean_as_target": mean_as_target.loc[models].values,
    "best_target_name": best_target_name.values, #as a source
    "best_target_value": best_target_value.values, #as a source
    "best_source_name": best_source_name.loc[models].values, #as a target
    "best_source_value": best_source_value.loc[models].values, # as a target
    "in_domain_performance": in_domain.values,
})

summary["diff_in_domain_minus_best_target"] = (
    summary["in_domain_performance"] - summary["best_target_value"]
)
summary["diff_in_domain_minus_mean_as_source"] = (
    summary["in_domain_performance"] - summary["mean_as_source"]
)
```
