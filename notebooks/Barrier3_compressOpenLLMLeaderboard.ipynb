```python
from glob import glob
from tqdm import tqdm
import pandas as pd

models = glob('../openLLMLeaderboard/*.pkl')

for model in tqdm(models):
    name = model.split('/')[-1]    
    df = pd.read_pickle(model)

    df = df[['doc_hash', 'input', 'correctness']]
    df.to_pickle(f'../openLLMLeaderboard/compressed/{name}')
```
