```python
import os

os.environ["HF_HOME"] = "/path/to/hf"
os.environ["HF_HUB_CACHE"] = "/path/to/hf"
os.environ["SENTENCE_TRANSFORMERS_HOME"] = "/path/to/hf"

import pandas as pd
from datasets import Dataset
from huggingface_hub import hf_hub_download
from sklearn.model_selection import train_test_split

fn = "routerbench_0shot.pkl" #"routerbench_raw.pkl", "routerbench_0shot.pkl", "routerbench_5shot.pkl"

path = hf_hub_download(
    repo_id="withmartian/routerbench",
    repo_type="dataset",
    filename=fn,
)

df = pd.read_pickle(path)
```

```python
models = ['WizardLM_WizardLM-13B-V1.2', 'claude-instant-v1', 'claude-v1', 'claude-v2', 'gpt-3.5-turbo-1106', 'gpt-4-1106-preview', 'meta_code-llama-instruct-34b-chat', 'meta_llama-2-70b-chat', 'mistralai_mistral-7b-chat', 'mistralai_mixtral-8x7b-chat', 'zero-one-ai_Yi-34B-Chat']
uneven = ['grade-school-math', 'consensus_summary', 'mtbench', 'mtbench-reference', 'mtbench-math', 'chinese_idioms']

df = df[~df['eval_name'].isin(uneven)] #exclude tasks with non-binary correctness label
df.columns = df.columns.str.replace('/', '_')
df = df[['sample_id', 'prompt', 'eval_name'] + models]

train, test = train_test_split(df, test_size=0.3, random_state=42) #like stated in the orignal paper

train.to_pickle('../routerBench/data/train.pkl')
test.to_pickle('../routerBench/data/test.pkl')
```
