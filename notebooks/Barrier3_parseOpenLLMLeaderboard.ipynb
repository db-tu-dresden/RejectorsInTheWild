```python
from glob import glob 
models = glob('/path/to/hf/datasets--open*/snapshots/*/*/*.json*')
models = [m for m in models if 'result' not in m]
```

```python
import os
from collections import defaultdict

def group_by_model(paths):
    groups = defaultdict(list)
    for p in paths:
        parts = p.split('/')
        model = parts[5]
        groups[model].append(p)
    return list(groups.values())

grouped = group_by_model(models)
```

```python
import pandas as pd
from tqdm import tqdm

for group in tqdm(grouped):
    dfs = []
    for file in group:
        df = pd.read_json(file, lines=True)
        
        temp = file.split('/')
        model = temp[-2]
        temp2 = temp[-1]
        temp3 = temp2.split('_')
        timestamp = ".".join(temp3[-1].split('.')[:-1])
        task = "_".join(temp3[1:-1])
        
        df['task'] = task
        df['timestamp'] = timestamp
        df['model'] = model
        dfs.append(df)
    df = pd.concat(dfs).reset_index()
    df.to_pickle(f'../openLLMLeaderboard/{model}.pkl')
```
